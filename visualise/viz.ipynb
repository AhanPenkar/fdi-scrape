```python
map_df = pd.read_csv('map_df.csv')
map_df.head()
```

```python
df3 = map_df.sort_values(by='Dollars', ascending=False)
df3.head()
```

```python
#remove the decimal from the dollars
df3['Dollars'] = df3['Dollars'].astype(int)
```

```python
Vodafone = df3[df3['Company'] == 'VODAFONE INDIA LTD']
Jio = df3[df3['Company'] == 'JIO PLATFORMS LIMITED']
Bosch = df3[df3['Company'] == 'BOSCH LIMITED']
Byju = df3[df3['Company'] == 'THINK & LEARN PRIVATE LIMITED']
One97 = df3[df3['Company'] == 'ONE97 COMMUNICATIONS LIMITED']
Amazon = df3[df3['Company'] == 'AMAZON SELLER SERVICES PVT LTD']
Arcelor = df3[df3['Company'] == 'ARCELORMITTAL INDIA PRIVATE LIMITED']
Axis = df3[df3['Company'] == 'AXIS BANK LIMITED']
Oyo = df3[df3['Company'] == 'ORAVEL STAYS PRIVATE LIMITED']
Airtel = df3[df3['Company'] == 'BHARTI AIRTEL LIMITED']
Flipkart = df3[df3['Company'] == 'FLIPKART INDIA PRIVATE LIMITED']
Amazon2 = df3[df3['Company'] == 'AMAZON SELLER SERVICES PRIVATE LIMITED']
Vodafone2 = df3[df3['Company'] == 'VODAFONE IDEA LIMITED']
Paytm = df3[df3['Company'] == 'ONE 97 COMMUNICATIONS LIMITED']

```

```python
#append all the dataframes
df_companies = pd.concat([Vodafone, Jio, Paytm, Flipkart, Byju, One97, Amazon, Arcelor, Axis, Oyo, Airtel, Amazon2, Vodafone2], axis=0).reset_index(drop=True)
df_companies.head(50)

```

```python
#group the data frames by company, country and sum the dollars
df_graph = df_companies.groupby(['Company', 'Country']).sum().reset_index()
df_graph.head(50)
```

```python
#show me unique countries in df_graph
df_graph['Country'].unique()
```

```python
#Replace HongKon with Hong Kong and save it in df_graph
df_graph['Country'] = df_graph['Country'].str.strip().replace('HongKon', 'Hong Kong')
df_graph['Country'] = df_graph['Country'].str.strip().replace('Virgin Islands(US)', 'Virgin Islands')
```

```python
df_graph['Country'].unique()
```

```python
df_graph.head(57)
```

```python
df_graph['Company'] = df_graph['Company'].str.strip().replace('AMAZON SELLER SERVICES PRIVATE LIMITED', 'AMAZON SELLER SERVICES PVT LTD')
df_graph['Company'] = df_graph['Company'].str.strip().replace('VODAFONE IDEA LIMITED', 'VODAFONE INDIA LTD')
df_graph['Company'] = df_graph['Company'].str.strip().replace('ONE 97 COMMUNICATIONS LIMITED', 'ONE97 COMMUNICATIONS LIMITED')
```

```python
#remove where dollars = 0
df_graph = df_graph[df_graph['Dollars'] != 0]
```

```python
df_graph.head(50)
```

```python
#GROUP BY COMPANY AND SUM THE DOLLARS and 
summy = df_graph.groupby(['Company']).sum().reset_index()
arrange = summy.sort_values(by='Dollars', ascending=False)
arrange
```

```python
#save to csv
df_graph.to_csv('df_graph.csv', index=False)
```
